```python
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_pickle("all_banks")
```

```python
df.head(5)
```

Create a separate dataframe for each bank 

```python
#*  Bank of America
#* CitiGroup
#* Goldman Sachs
#* JPMorgan Chase
#* Morgan Stanley
#* Wells Fargo

BAC = df["BAC"]
C = df["C"]
GS = df["GS"]
JPM = df["JPM"]
MS = df["MS"]
WFC = df["WFC"]
```

Create a list of the ticker symbols as strings in alphabetical order

```python
tickers = ["BAC", "C", "GS", "JPM", "MS", "WFC"]
```

** Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], keys = tickers, axis = 1)
```

```python
# a multilevel index is created
bank_stocks.head()
```

Set the column name levels

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    
    returns[tick + "Return"] = bank_stocks[tick]["Close"].pct_change()
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
BAC.head()
```

```python
plt.figure(figsize = (12, 4))
plt.plot(BAC.loc["2008-01-01" : "2008-12-31"]["Close"].rolling(window = 30).mean(), color = "#0097e6", 
         label = "30 moving avrg")
plt.plot(BAC.loc["2008-01-01" : "2008-12-31"]["Close"], color = "#e1b12c", label = "close price BAC 2008")
plt.grid(True)
plt.legend()
plt.show()
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
bank_stocks.xs(key = "Close", axis = 1, level = "Stock Info").corr()
correlation = bank_stocks.xs(key = "Close", axis = 1, level = "Stock Info").corr()
plt.figure(figsize = (10, 5))
sns.heatmap(correlation, cmap = "coolwarm", annot = True)
```

```python
correlation.iplot(kind = "heatmap", colorscale = "rdylbu")
```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(correlation, annot = True, cmap = "coolwarm")
```
