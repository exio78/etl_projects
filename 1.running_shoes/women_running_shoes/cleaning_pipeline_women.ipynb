With regular expressions we can separate the text of three technologies. I had lots of problems figuring out this.

```python
def midsole(midsole_sentence):
    
    midsole_sentence = re.findall("MIDSOLE(.+)OUTSOLE", midsole_sentence)
    if len(midsole_sentence) == 0:
        return "no information"
    else: 
        return midsole_sentence[0].replace("\xa0", "")
           
df["technology"].apply(midsole)
```

```python
def outsole(outsole_sentence):
    
    outsole_sentence = re.findall("OUTSOLE(.+)UPPER", outsole_sentence)
    if len(outsole_sentence) == 0:
        return "no information"
    else: 
        return outsole_sentence[0].replace("\xa0", "")
           
df["technology"].apply(outsole)
```

```python
def upper(upper_sentence):
    
    upper_sentence = re.findall("UPPER(.+)", upper_sentence)
    if len(upper_sentence) == 0:
        return "no information"
    else: 
        return upper_sentence[0].replace("\xa0", "")
           
df["technology"].apply(upper)
```

```python
df["midsole"] = df["technology"].apply(midsole)
df["outsole"] = df["technology"].apply(outsole)
df["upper"] = df["technology"].apply(upper)
```
