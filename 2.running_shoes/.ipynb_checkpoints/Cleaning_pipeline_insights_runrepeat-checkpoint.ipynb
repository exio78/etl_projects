# Separate the price column into three, the price, the discount and the discount in percentage.

```python
df_women.loc[0, "Price"] # we can use regular expressions to create a list of three values per row
```

 with regular expressions the pattern \d+ is designed to extract sequence of digits

```python
digits = re.findall("\d+",df_women.loc[0, "Price"])
```

```python
digits
```

```python
df_women["Price"].apply(lambda x : re.findall("\d+", x))
```

```python
df_women["Price"] = df_women["Price"].apply(lambda x : re.findall("\d+", x)) #update the dataframe
```

```python
df_women
```

**Now we need to create a pandas series, a list of lists to replace the current column price by the new column price, because as you can see we have in the price column some empty lists and we need to match the proportions of lenght of those lists to be replaced on column from another.**

**Once we have the column replaced with a column where each row has a list with three items, we can apply a lambda function with the apply method to create the three separated and desired columns, pirce, price discount and price with %. We need to select something for the new columns, we can't select an empty list to create the value of a new column**

```python
# create a list of lists where empty placeholders will be equal to 0
new_column = []
for i in df_women["Price"]:
    
    if i == []:
        i = [0, 0, 0]
    if len(i) == 1:
        i = [i[0], 0, 0]
    new_column.append(i)
    print(i)
```

```python
# we convert this new list of list into a pandas series in order to be replaced by the original column
new_column = pd.Series(new_column) 
new_column
```

```python
df_women["Price"] # this is the original column
```

```python
df_women["Price"] = new_column # update the column 
```

```python
df_women["Price"]
```

```python
# now we can select the index of the item of each list of the 
#price column to select the desired value for the new columns.

df_women["price"] = df_women["Price"].apply(lambda x : x[0])
df_women["price_discount"] = df_women["Price"].apply(lambda x : x[1])
df_women["price_discount_%"] = df_women["Price"].apply(lambda x : x[2])

df_women
```

```python
df_women.drop("Price", axis = 1, inplace = True) # we don't need the price original column
```

```python
df_women
```

# We need to clean the Title column, create the columns for brand and model of the pair of shoes.

**We are going to use the split method and a lambda function with apply to select the first index to create the new brand column, in the cases where we have two or more words for the brand we are going to replace those words with a replace function using a dictionary to complete the name of the brand.** 

**We are going to apply the same idea for the model, but instead of selecting the index 0 we are going to select from index 1 towards the rest of indexes, and for those cases where we have extra text we are going to replace them with nothing replace(text, "")**

```python
df_women["Title"][0].split(" ")
```

```python
df_women["Title"].str.split(" ")
```

```python
df_women["Title"].str.split(" ").apply(lambda x : x[0])
```

```python
# set the brand column
df_women["brand"] = df_women["Title"].str.split(" ").apply(lambda x : x[0])
```

Here we can see how many brands we have, 38 brands and how many shoes per brand.

```python
df_women["brand"].value_counts()
```

```python
df_women["brand"].nunique()
```

```python
# we can create the dictionary to complete the name of the brand for those 
# running shoes which have more than one word on their brands.
# apparently we have 8 brands with more than one word.
brands_complete = {"New" : "New Balance", 
                   "Under": "Under Armour", 
                   "La": "La Sportiva", 
                   "Topo" : "Topo Athletic",
                   "The": "The North Face",
                   "Xero": "Xero Shoes",
                   "Vibram": "Vibram FiveFingers",
                   "APL": "APL Performance"}
```

```python
df_women["brand"].replace(brands_complete, inplace = True)
```

```python
df_women
```

**Now it is time to create the model column.**

```python
# we are going to sleect everything from index 1 and replace the extra text 
# by nothing in the case of those brands with more than one word
df_women["model"] = df_women["Title"].str.split(" ").apply(lambda x : x[1:])

df_women["model"] = df_women["model"].apply(lambda x : " ".join(x))

df_women["model"] = df_women["model"].apply(lambda x : x.replace("Balance", ""))
df_women["model"] = df_women["model"].apply(lambda x : x.replace("Armour", ""))
df_women["model"] = df_women["model"].apply(lambda x : x.replace("Sportiva", ""))
df_women["model"] = df_women["model"].apply(lambda x : x.replace("Athletic", ""))
df_women["model"] = df_women["model"].apply(lambda x : x.replace("North Face", ""))
df_women["model"] = df_women["model"].apply(lambda x : x.replace("Shoes", ""))
df_women["model"] = df_women["model"].apply(lambda x : x.replace("FiveFingers", ""))
df_women["model"] = df_women["model"].apply(lambda x : x.replace("Performance", ""))
```

```python
df_women
```

# # **% of running shoes that fall into different price ranges with the webpage Runrepeat**

We are going to take in consideration prices that are different to 0 because remember, 0 on this webpage means out of stock, if we analyse this data with 0 in it, this can affect the percentage of price range for all of this data set. However, 0 is replaced by None 

```python
df_women["price"]
```

```python
price_range = df_women[df_women["price"] > 0.0] # we can use filter too
```

```python
price_range # we have reduced our dataset from 1199 to 1087
```

```python
max(price_range["price"])
```

```python
min(price_range["price"]) # it is not 0 because we selected non zero values
```

So the range of prices are between 45 and 419 dollars, let's create 5 ranges splitting for each 75 dollars with a pie chart.

```python
filt_75 = price_range["price"] <= 75
price_range[filt_75]["price"].count()
```

```python
filt_75_150 = (price_range["price"] > 75) & ((price_range["price"] <= 150))
price_range[filt_75_150]["price"].count()
```

```python
filt_150_225 = (price_range["price"] > 150) & ((price_range["price"] <= 225))
price_range[filt_150_225]["price"].count()
```

```python
filt_225_330 = (price_range["price"] > 225) & ((price_range["price"] <= 330))
price_range[filt_225_330]["price"].count()
```

```python
filt_330_419 = (price_range["price"] > 330) & ((price_range["price"] <= 419))
price_range[filt_330_419]["price"].count()
```

```python
one = price_range[filt_75]["price"].count()
two = price_range[filt_75_150]["price"].count()
three = price_range[filt_150_225]["price"].count()
four = price_range[filt_225_330]["price"].count()
five = price_range[filt_330_419]["price"].count()

fig = plt.figure()
ax = fig.add_subplot()

five_parts = [one, two, three, four, five]
labels = ["45-75$", "75-150$", "150-225$", "225-330$", "330-419$"]
colours = ["#00d2d3", "#2e86de", "#ff9f43", "#ee5253", "#5f27cd"]

ax.pie(five_parts, colors = colours, autopct ="%1.2f%%", pctdistance = 1.17, wedgeprops= {"edgecolor": "#222f3e"})

ax.set_title("Price Ranges Running Shoes")
ax.legend(labels, loc = [0.9, 0.05])
fig.set_facecolor("#c8d6e5")
plt.show()
```
